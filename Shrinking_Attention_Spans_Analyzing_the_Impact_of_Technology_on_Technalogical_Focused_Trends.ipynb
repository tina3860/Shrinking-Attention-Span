**Has technology &  media consumption impacted the human attention span?**

Have you ever heard the phrase, “the attention span of a goldfish”? It seems increasingly relevant today as human attention spans shrink, partly due to the influence of digital technology. Attention span can be defined by three components: direction, depth, and duration.

## Direction & Depth

Have you ever found yourself going down a “rabbit hole” after encountering just one post? You start with a single video, then transition to articles, and before you know it, you end up on an e-commerce site. This process illustrates the flow of tasks and subjects we encounter online. One of Microsoft’s Adoption and Change Specialists offers an analogy: think of our attention as a flashlight. It shines a beam on different topics and tasks at various moments. Now, consider the challenge of trying to point your flashlight in multiple directions at once. Is that an effective way to search for something or to stay on course? This leads to the question of whether attention switching or multitasking is productive, especially in terms of work and our overall attention span. Digital media plays a significant role in this discussion; it encourages multitasking through constant notifications and pop-up ads that divert our focus between apps, messages, and tasks.

Depth relates to how engaged we are in a task—essentially, how immersed we feel. This is about making a connection with what we are doing. Have you ever been so engrossed in what you're doing that you lose track of time until the day is over? Technology, our phones as a big example, interrupts the ability to connect deeply and focus on tasks at hand. How many times have you pulled out your phone, even to just look at the time on the lock screen, during a meeting or class? In today’s media landscape, constant disruptions through notifications, social media and multi-app interactions disrupt the immersive state and our “flow”.

```python
#create dict
columns_dict = {
    'checks': [f'Day{i}TotalChecks' for i in range(1, 14)],
    'hours': [f'Day{i}HoursUse' for i in range(1, 14)]
}

checks = pd.DataFrame(columns=columns_dict['checks'])
hours = pd.DataFrame(columns=columns_dict['hours'])
```

```python
#create age groups
bins = [18, 20, 25, 30, 35]
labels = ['Young Adults \n(18-20)', 'Early 20s \n(21-25)', 'Late 20s \n(26-30)', 'Early 30s \n(31-35)']

# Create a new column for Age Group
typical_smartphone['AgeGroup'] = pd.cut(typical_smartphone['Age'], bins=bins, labels=labels, right=False)


#avg number of device checks
typical_smartphone['TotalChecks'] = typical_smartphone[columns_dict['checks']].mean(axis=1)
typical_smartphone['TotalChecks']

#no particp. in each age group
age_group_participants = typical_smartphone['AgeGroup'].value_counts().sort_index()
age_group_participants

#avg no hours use
typical_smartphone['TotalHoursUse'] = typical_smartphone[columns_dict['hours']].mean(axis=1)
typical_smartphone['TotalHoursUse']

#Sum the total checks and hrs by age group
age_group_checks = typical_smartphone.groupby('AgeGroup')['TotalChecks'].sum().sort_index()
age_group_hrs = typical_smartphone.groupby('AgeGroup')['TotalHoursUse'].sum().sort_index()

# plt
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(18, 5))
ax1.bar(age_group_checks.index, age_group_checks.values)
ax1.set_title('Avg Number of Device Checks by Age Group')
ax1.set_xlabel('Age Group')
ax1.set_ylabel('Avg Number of Checks')

ax2.bar(age_group_hrs.index, age_group_hrs.values)
ax2.set_title('Avg Number of Hours Used by Age Group')
ax2.set_xlabel('Age Group')
ax2.set_ylabel('Avg Number of Hours')


ax3.bar(age_group_participants.index, age_group_participants.values)
ax3.set_title('Number of Participants by Age Group')
ax3.set_xlabel('Age Group')
ax3.set_ylabel('Number of Participants')

fig.suptitle('Comparison of Average Checks and Usage by Age Group', fontsize=16)
plt.show()
```

**Analyze Song Duration Trends**

From the early 1960s to the late 1970s, there is a noticeable upward trend in average song length. The average song length rises from about 2 minutes in the early 1960s to over 5 minutes in the early 1970s. This trend could reflect a shift in music production and listener preferences toward more extended compositions, possibly influenced by genres like rock and progressive music, which often included longer instrumental sections. For example, the famous rock band, Guns and Roses, are known for their rock ballads that consists of longer instrumentals.

From the late 1970s through the early 2000s, the graph shows considerable fluctuation. The average song length oscillates between 4 and 5 minutes, without a clear upward or downward trend. This period likely reflects varied music industry trends, with shifts across genres and listener preferences. While some genres, like pop, tended to have shorter songs, others like rock and jazz often had longer tracks.

Around the early 2000s, there is a noticeable dip in the average song length, with some years showing averages close to or below 4 minutes.
From the 2010s onwards, song lengths seem to stabilize at around 4 minutes, with minor fluctuations. This could be influenced by the digital age and the rise of streaming platforms, where shorter, catchier songs perform better for audience retention and playlist inclusion. The rise of the internet, streaming platforms, and social media in the 2000s and 2010s likely influenced song lengths. As attention spans may have shortened, artists and producers might have shifted toward creating shorter, more impactful songs that fit the “instant gratification” culture encouraged by digital media.

Different musical eras often emphasize different types of songs and genres. For instance, the progressive and rock movements in the 1970s favored longer songs with extended solos and instrumental sections, while the pop and hip-hop trends in the 2000s and 2010s often focused on shorter, more radio-friendly tracks. The general stabilization or even slight reduction in song length post-2000 may indicate that artists and record labels are increasingly catering to shorter attention spans, particularly with the rise of short-form content on platforms like TikTok and Instagram influencing music trends.

```python
#avg song length per year
avg_duration_per_year = spotify.groupby('Year')['Duration_Mins'].mean()

#plot
avg_duration_per_year.plot(kind='line', color = 'green')
plt.xlabel('Year')
plt.ylabel('Avg Song Length (in mins)')
plt.title('Average Song Length per Year')
```

**Compare with Popularity Metrics**

Popularity metrics are on a scale from 1-100, the higher the rating, the more popular the song is.

The majority of songs with high popularity scores (above 70) tend to have durations between 2 to 5 minutes. This indicates that songs in the shorter duration range are more likely to be popular, aligning with the general preference for concise, engaging content in the music industry.

As song duration increases past 5 minutes, there is a noticeable decline in popularity scores. Very few songs with durations over 10 minutes reach a popularity score above 60. This suggests that longer songs generally do not perform as well in terms of popularity, which could be due to shorter attention spans or listener preferences for shorter, easily consumable music.

The clustering of high-popularity songs within the 2 to 5-minute range may reflect shifting audience attention spans. With the rise of platforms like TikTok and streaming services, listeners may prefer songs that are shorter and provide a quick, engaging experience. Coming from personal experience and observations, a part of the listening population can't even get through a 3 minute song, let alone a song past 5 minutes.
